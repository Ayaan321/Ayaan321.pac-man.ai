```python
def create_board(rows, cols, n_row, n_col, p_row, p_col):
    board = []
    for row in range(rows):
        curr_row = []
        for col in range(cols):
            curr_row.append("0")
        board.append(curr_row)
    board[p_row][p_col] = "\U0001F606"
    board[n_row][n_col]= "\U0001f44d"
    return board
```

```python
def printer(board):
    for row in board:
        print(row)
```

```python
board = create_board(4, 4, 2, 3, 3, 2) 
printer(board)
```

```python

def bfs(board, rows, cols, p_row, p_col):
    max_row = rows
    max_col = cols
    queue = [[p_row, p_col]]
    while len(queue) > 0:
        elem_list = queue.pop(0)
        row = elem_list[0]
        col = elem_list[1]
        
        if board[row][col] == "\U0001f44d":
            print("found at", row, col)
            printer(board)
            return
        
        board[row][col]= "\U0001F606"
        up_row, up_col = row-1, col
        if up_row >= 0:
            queue.append([up_row,up_col])
        down_row, down_col = row+1, col
        if down_row < max_row:
            queue.append([down_row,down_col])
        right_row, right_col = row,col+1
        if right_col < max_col:
            queue.append([right_row,right_col]) 
        left_row, left_col = row, col-1
        if left_col >= 0,
            queue.append([left_row,left_col])
    print(board)
```

```python
num_rows = int(input("enter the number of rows: "))
num_cols = int(input("enter the number of columns: "))
pac_row = int(input("enter the row of pac man: "))
pac_col = int(input("enter the column of pac man  "))
end_row = int(input("enter the end row "))
end_col = int(input("enter the end column"))
board = create_board(num_rows, num_cols, end_row, end_col, pac_row, pac_col,) 
printer(board)
bfs(board, num_rows, num_cols, pac_row, pac_col )
```
